Q:What is the difference between the Pandas series and dataframe?

pandas series
*  Data of any kind (integer, float, text, etc.) can be stored in a one-dimensional array-like object called a Pandas Series. Each element has a distinct identification known as an index since it is labeled. A series can be compared to a single column in a database table or a column in a spreadsheet. A basic data structure in Pandas, series are frequently employed for tasks involving data analysis and manipulation. Lists, arrays, dictionaries, and pre-existing Series objects can all be used to build them. Additionally, series serve as a foundation for the more intricate Pandas DataFrame, a two-dimensional table-like structure made up of several Series objects.
*   An index that corresponds to each element in a series can be used to read or modify the data
*   Because Series enables vectorized operations, you can effectively perform arithmetic operations on the entire serie
*   Pandas automatically aligns the data based on the index labels when executing operations between two Series objects.
*  In series operations, missing values, denoted by NaN (Not a Number), can be handled elegantly.








pandas DataFrame
*   A two-dimensional, tabular data structure with rows and columns is called a Pandas DataFrame. It resembles a table in a relational database or a spreadsheet. The data, which is stored in rows and columns, the rows, which are labeled by an index, and the columns, which are labeled and contain the actual data, are the three primary parts of the DataFrame.
*   DataFrame provides flexible indexing options, allowing access to rows, columns, or individual elements based on labels or integer positions.
*   Columns in a DataFrame are Series objects, enabling various operations such as arithmetic operations, filtering, and sorting.
*   DataFrames provide methods for handling missing or NaN values, including dropping or filling missing values.
* DataFrames support group-by operations for summarizing data and applying aggregation functions.



# **Difference**
# 1.    ***Series***
*   One- dimensional
*   Series elements must be homogenous.
*   Immutable(size cannot be changed).
*   Element wise computations.
*   Functionality is less
*   Alignment not supported.


# 2.   ***DataFrame***
*  Two- dimensional
*   Can be heterogeneous.
*   Mutable(size can be changeable).
*   Column wise computations.
*   More functionality
*  Alignment is supported.
