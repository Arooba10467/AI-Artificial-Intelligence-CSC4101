# Part 2

● Use pytorch

```python
import torch
```

```python
import numpy as np
import pandas as pd
```

● Use Generated dataset from Assignment 3

```python
train_df = pd.read_csv("train_data_A3.csv")
test_df = pd.read_csv("test_data_A3.csv")

X_train = torch.tensor(train_df.iloc[:, :-1].values, dtype=torch.float32)
y_train = torch.tensor(train_df.iloc[:, -1].values, dtype=torch.float32).unsqueeze(1)

X_test = torch.tensor(test_df.iloc[:, :-1].values, dtype=torch.float32)
y_test = torch.tensor(test_df.iloc[:, -1].values, dtype=torch.float32).unsqueeze(1)
```

● Design a Logistic Regression using Pytorch

```python
input_size = X_train.shape[1]

# weights
w = torch.zeros((input_size, 1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

● Print model weights before training


```python
print("Initial Weights:", w)
print("Initial Bias:", b)
```

● Train the model on Generated training data for 10 epochs
* Loss must be printed during training to show how training is gong on

```python
lr = 0.01
epochs = 10

losses = [] # Initialize a list to store losses

for epoch in range(epochs):

    # Forward pass
    z = torch.matmul(X_train, w) + b
    y_pred = torch.sigmoid(z)

    # Loss (Binary Cross Entropy)
    loss = -torch.mean(y_train * torch.log(y_pred) + (1 - y_train) * torch.log(1 - y_pred))

    print(f"Epoch {epoch+1}, Loss: {loss.item()}")
    losses.append(loss.item()) # Append the loss to the list

    # Backpropagation
    loss.backward()

    # Update weights
    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad

    # reset gradients
    w.grad.zero_()
    b.grad.zero_()
```

● Print new weights


```python
print("Final Weights:", w)
print("Final Bias:", b)
```

● Make predictions on testing dataset


```python
with torch.no_grad():
    z = torch.matmul(X_test, w) + b
    y_pred = torch.sigmoid(z)

    y_pred_class = (y_pred >= 0.5).float()
```

● Evaluate results using Evaluation metric (what evaluation metric we use for logistic
regression)

```python
accuracy = (y_pred_class == y_test).float().mean()

print("Accuracy:", accuracy.item())
```
